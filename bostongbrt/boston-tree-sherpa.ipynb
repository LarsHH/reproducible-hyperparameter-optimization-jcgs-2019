```python
import sherpa
```

```python
parameters = [sherpa.Continuous('lr_scaler', [0.5, 1.5]),
              sherpa.Discrete('n_estimators', [100, 400]),
              sherpa.Discrete('max_depth', [2, 10]),
              sherpa.Continuous('subsample', [0.5, 1.])]
```

```python
algorithm = sherpa.algorithms.RandomSearch(max_num_trials=1000, repeat=25)        
```

```python
scheduler = sherpa.schedulers.LocalScheduler()
```

```python
sherpa.optimize(algorithm=algorithm,
                scheduler=scheduler,
                parameters=parameters,
                lower_is_better=True,
                filename="boston-tree.py",
                max_concurrent=10,
                output_dir='./output_boston_regressor')
```
